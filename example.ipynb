```julia
using pflow: Pflow, place!, arc!, guard!, transition!, to_json, to_svg, to_html, to_model, to_state, default_rates

function knapsack!(m::Pflow)
    # Add places
    place!(m, "item0", 0, 1, nothing, 351, 140)
    place!(m, "item1", 1, 1, nothing, 353, 265)
    place!(m, "item2", 2, 1, nothing, 351, 417)
    place!(m, "item3", 3, 1, nothing, 350, 543)
    place!(m, "weight", 4, 0, nothing, 880, 320)
    place!(m, "value", 5, 0, nothing, 765, 145)
    place!(m, "capacity", 6, 15, nothing, 730, 541)

    transition!(m, "txn0", 0, "", 465, 139)
    transition!(m, "txn1", 1, "", 466, 264)
    transition!(m, "txn2", 2, "", 462, 418)
    transition!(m, "txn3", 3, "", 464, 542)

    # Add arcs
    arc!(m, "txn0", "weight", 2)
    arc!(m, "txn0", "value", 10)
    arc!(m, "txn1", "weight", 4)
    arc!(m, "item0", "txn0", 1)
    arc!(m, "txn1", "value", 10)
    arc!(m, "item1", "txn1", 1)
    arc!(m, "item2", "txn2", 1)
    arc!(m, "item3", "txn3", 1)
    arc!(m, "txn2", "weight", 6)
    arc!(m, "txn2", "value", 12)
    arc!(m, "txn3", "value", 18)
    arc!(m, "txn3", "weight", 9)
    arc!(m, "capacity", "txn0", 2)
    arc!(m, "capacity", "txn1", 4)
    arc!(m, "capacity", "txn2", 6)
    arc!(m, "capacity", "txn3", 9)
end

m = Pflow()
knapsack!(m)
display(HTML(to_html(m)))
```

```julia
using Plots
using Petri
using LabelledArrays
using Plots
using OrdinaryDiffEq


petri_net = to_model(m)

# Plot the Petri net graph
graph = Graph(petri_net)

time_max = 5.0
tspan = (0.0, time_max)

# Display graphviz graph
# display(graph)

rates = default_rates(m)
initial_state = to_state(m)
prob = ODEProblem(petri_net, initial_state, tspan, rates)
sol = solve(prob, Tsit5())
(rates=rates, state=initial_state)
```

```julia
# Plot results
plot(title="Knapsack Simulation", xlabel="Time", ylabel="Tokens", lw=2)
plot!(sol.t, sol[:value, :], label="Value")

value = round(sol[:value, end], digits=3)
annotate!(4, value, text("Value: $value", 10, :left))

# Display the plot no legend
plot!(legend=false)
```
